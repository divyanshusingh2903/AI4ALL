```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./hospital_readmissions.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# DATA Analysis
import matplotlib.ticker as mticker

categorial_variables = ['age', 'medical_specialty', 'diag_1', 'diag_2', 'diag_3',
       'glucose_test', 'A1Ctest', 'change', 'diabetes_med']

# creating figure with subplots
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(15,15))
axs = axs.flatten()


fig.suptitle("Countplot of Categorical Variables")

for i,var in enumerate(categorial_variables):
    sns.countplot(x=var, hue='readmitted', data=df, ax=axs[i])
    axs[i].set_xticklabels(axs[i].get_xticklabels(), rotation=90)

fig.tight_layout()    
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['readmitted'], axis=1)
y = df['readmitted']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
train_data = x_train.join(y_train)
train_data
```

```python
train_data.hist(figsize=(20,20))
```

```python
# turn boolean data to int (true = 1 and false = 0)
train_data['readmitted'] = train_data['readmitted'] == 'yes'
train_data['diabetes_med'] = train_data['diabetes_med'] == 'yes'
train_data['change'] = train_data['change'] == 'yes'
```

```python
value_mapping = {
    'no': 0,
    'normal': 1,
    'high': 2
}

train_data['A1Ctest'] = train_data['A1Ctest'].map(value_mapping)
train_data['glucose_test'] = train_data['glucose_test'].map(value_mapping)
```

```python
train_data['diag_1'].unique()
```

```python
diag_mapping = {
    'Circulatory': 0,
    'Respiratory': 1,
    'Other': 2,
    'Digestive': 3,
    'Injury': 4,
    'Musculoskeletal': 5,
    'Diabetes': 6,
    'Missing': 7
}

train_data['diag_1'] = train_data['diag_1'].map(diag_mapping)
train_data['diag_2'] = train_data['diag_2'].map(diag_mapping)
train_data['diag_3'] = train_data['diag_3'].map(diag_mapping)
```

```python
train_data
```

```python
train_data['age'].unique()
```

```python
age_mapping = {
    '[70-80)': 80,
    '[50-60)': 60,
    '[60-70)': 70,
    '[40-50)': 50,
    '[80-90)': 90,
    '[90-100)': 100
}

train_data['age'] = train_data['age'].map(age_mapping)
```

```python
train_data = train_data.drop(['medical_specialty'], axis=1)
train_data
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
```
